```python
food=driver.find_elements_by_class_name('_2_M76')
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
for i in food:
    nme=i.get_attribute('innerHTML')
    if '#' in nme:
        continue
    print(remove_html_tags(nme))
#getting a list of all tyhe handles that come after writing food 
#and removing the ones having # in it using remove_html_tags function
```
